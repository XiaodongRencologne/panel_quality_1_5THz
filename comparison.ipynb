```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x,y=np.meshgrid(np.linspace(-0.001,0.001,501)*180/np.pi*3600,np.linspace(-0.001,0.001,501)*180/np.pi*3600)
ref=np.genfromtxt('output/infocus_far/source_field.txt',delimiter=',')
ref=ref[...,0]+1j*ref[...,1]
ref=ref.reshape(501,-1)

test=np.genfromtxt('output/infocus_far_panelerror/panel_error_effect1500GHz.txt',delimiter=',')
test=test[...,0]+1j*test[...,1];
test=test.reshape(501,-1)

test2=np.genfromtxt('output/infocus_far_panel_corrected/panel_error_after_correction_1500GHz.txt',delimiter=',')
test2=test2[...,0]+1j*test2[...,1];
test2=test2.reshape(501,-1)
```
